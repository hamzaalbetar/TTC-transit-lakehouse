###Bronze ingestion - routes
###Source: TTC
###Layer: Bronze

##Reading file from catalog

```python
df = spark.read.option("header", "true").csv("/Volumes/ttc-lakehouse/raw/ttc/routes.txt")
```

##write into bronze layer

```python
df.write.format("delta").mode("overwrite").saveAsTable("`ttc-lakehouse`.bronze.routes")
```
