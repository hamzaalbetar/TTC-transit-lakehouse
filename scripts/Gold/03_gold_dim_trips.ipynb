#read table from silver layer

```python
df = spark.read.format("delta").table("`ttc-lakehouse`.silver.trips")
```

#Join trps with route dimension to enrich trip-level data with route attributes

```python
df.createOrReplaceTempView("trips_temp")

query = """
SELECT 
t.trip_id,
t.route_id,
t.service_id,
t.direction_id,
t.shape_id, 
r.route_short_name, 
r.route_type
FROM trips_temp t
JOIN `ttc-lakehouse`.gold.dim_routes r
ON t.route_id = r.route_id
"""

df = spark.sql(query)
```

##Write into Gold layer

```python
df.write.format("delta").mode("overwrite").saveAsTable("`ttc-lakehouse`.gold.dim_trips")
```
