#read table from silver layer

```python
df = spark.read.format("delta").table("`ttc-lakehouse`.silver.stop_times")
```

##select relevent attributes for building the gold stop times fact 

```python
df = df.select('trip_id','stop_id','arrival_time','departure_time','stop_sequence')
```

##write into Gold layer

```python
df.write.format("delta").mode("overwrite").saveAsTable("`ttc-lakehouse`.gold.fact_stop_times")
```
