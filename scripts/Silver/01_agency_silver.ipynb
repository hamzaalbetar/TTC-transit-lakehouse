#read table from bronze layer

```python
df = spark.read.format("delta").table("`ttc-lakehouse`.bronze.agency")
```

##make colomn name friendly

```python
df = df.withColumnRenamed("agency_lang", "agency_language")
```

##make language name friendly

```python
from pyspark.sql.functions import when

df = df.withColumn(
    "agency_language",
    when(df["agency_language"] == "en", "English").otherwise(df["agency_language"])
)
```

##flag standardization (0/1 -> Boolean)

```python
df = df.withColumn(
    "cemv_support",
    when(df["cemv_support"] == 1, True).when(df["cemv_support"] == 0, False).otherwise(df["cemv_support"])
)
```

##write into silver layer


```python
df.write.format("delta").mode("overwrite").saveAsTable("`ttc-lakehouse`.silver.agency")
```
