#read table from bronze layer

```python
df = spark.read.format("delta").table("`ttc-lakehouse`.bronze.calendar_dates")
```

%md
##Data type standardization : cast date from String to Date

```python
from pyspark.sql.functions import try_to_date, regexp_replace, col


df = df.withColumn("date", try_to_date(regexp_replace(col("date"), r"(\d{4})(\d{2})(\d{2})", r"$1-$2-$3"), "yyyy-MM-dd"))
```

##write into silver layer

```python
df.write.format("delta").mode("overwrite").saveAsTable("`ttc-lakehouse`.silver.calendar_dates")
```
